'''
Data Manipulation using pandas
Merging and Joining dataframes with pandas
'''

```python
class_rankings ={'Stud Name':['Chaitanya','Manu','Seema','Vaman'], 'Rank':[1,2,3,4]}
```

```python
class_rankings
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(class_rankings)
```

```python
df1
```

```python
majors = {'Stud Name': ['Chaitanya','Manu','Seema','Vaman'], 'Majors':['Computer Science','Electronics','Arts','Biotech']}
```

```python
df2 = pd.DataFrame(majors)
```

```python
df2
```

```python
print(pd.merge(df1,df2,on='Stud Name'))
```

```python
print(pd.merge(df1,df2,on='Stud Name',how='left'))
```

```python
majors1 = {'Stud Name': ['Chaitanya','Manu','Seema','Vaman','chaitanya2'], 'Majors':['Computer Science','Electronics','Arts','Biotech','Data Analytics']}
```

```python
majors1
```

```python
df3 = pd.DataFrame(majors1)
```

```python
df3
```

```python
print(pd.merge(df1,df3,on = 'Stud Name', how='left'))
```

```python
print(pd.merge(df1,df3,on = 'Stud Name', how='right'))
```

```python
print(pd.merge(df1,df3,on = 'Stud Name', how='outer'))
```

```python
print(pd.merge(df1,df3,on = 'Stud Name', how='inner'))
```
